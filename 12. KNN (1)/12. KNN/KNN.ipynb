```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head(25)
```

```python
df.info()
```

```python
df1=df.drop(columns='Unnamed: 32')
```

```python
df1.head()
```

```python
df1.diagnosis.value_counts()
```

```python
df1['diagnosis']=np.where(df1['diagnosis']=='M',1,0)
```

```python
df1.head()
```

```python
x=df1.drop(columns=['id','diagnosis'])
```

```python
x.head()
```

```python
y=df1.diagnosis
```

```python
y.head()
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(xtrain, ytrain)
```

```python
y_pred = knn.predict(xtest)
```

```python
y_pred
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ytest,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import roc_auc_score,roc_curve,auc
```

```python
false_positive_rate,true_positive_rate,thresholds=roc_curve(ytest,y_pred)
roc_auc=auc(false_positive_rate,true_positive_rate)
roc_auc
```

```python
roc_auc_score(ytest,y_pred)
```

```python
plt.plot(false_positive_rate,true_positive_rate)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.show()
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
classification_report(ytest,y_pred)
```
